**Ejercicio 1:** Traducí a Python tu solución al ejercicio 9 de la lección 5 (https://studio.code.org/s/20-hour/stage/5/puzzle/9) y agregale lo necesario para que realice el siguiente dibujo

image.png

Para ello, observá que:

*   La circunferencia que se dibujó mide ```360```.
*   Por lo tanto su diámetro mide 360/pi y su radio 180/pi
*   Podés escribir instrucciones como `forward(360/math.pi)`, también se puede usar `penup()` para desplazarse sin dibujar y luego `pendown()` para volver a dibujar.

Podés ejecutar:

````
# Escribí tu código abajo de esta linea

# Instalación de ColabTurtle e importación de sus funciones
!pip3 install ColabTurtle
from ColabTurtle.Turtle import *

# Configuración de pizarra inicial 
initializeTurtle()
bgcolor("white")

# Solución del ejercicio 9 de la lección 5
color("black")
speed(13)
for _ in range(360):
    forward(1)
    right(1)
# Completar dibujo
penup()
forward(58)
pendown()
color("blue")
speed(10)
for _ in range(4):
    right(90)
    forward(115)
````

**Ejercicio 2:** Traducí a Python tu solución al ejercicio 8 de la lección 7 (https://studio.code.org/s/20-hour/stage/7/puzzle/8).

````
# Escribí tu código abajo de esta linea

# Configuración pizarra inicial
initializeTurtle()
bgcolor("white")

# Elección aleatoria de color
from random import choice
choice(["yellow", "orange", "red", "green", "blue", "purple", "grey", "black"])

# Solución del ejercicio 8 de la lección 7
speed(13)
for _ in range(4):
    for _ in range(10):
        color(choice(["yellow", "orange", "red", "green", "blue", "purple", "grey", "black"]))
        for _ in range(4):
            forward(20)
            right(90)
        forward(20)
    right(90)
````
